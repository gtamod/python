```python
def sinplot(flip=1): 
    x = np.linspace(0, 15, 100) 
    for i in range(1, 5): 
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip) 
sns.set_style("white")
sns.set_palette("husl") 
sinplot() 
plt.show()
```

```python
#for conditional relationships
g = sns.FacetGrid(tips, col="time",  row="smoker")
g = g.map(plt.hist, "total_bill")
plt.show()
```

```python
#for pairwise relationships
g = sns.PairGrid(ds)
g = g.map(plt.scatter)
plt.show()
```

```python
#univariate distribution on diagonal
g = sns.PairGrid(ds)
g = g.map_diag(plt.hist)
g = g.map_offdiag(plt.scatter)
```
