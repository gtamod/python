```python
import matplotlib.pyplot as plt
days=[1,2,3,4,5]
sleeping=[7,8,6,11,17]
eating=  [2,3,4,3,2]
working= [7,8,7,2,2]
playing= [8,5,7,8,13]
plt.plot([],[],color='m',label='sleeping', linewidth=5)
plt.plot([],[],color='c',label='eating',   linewidth=5)
plt.plot([],[],color='r',label='working',  linewidth=5)
plt.plot([],[],color='k',label='playing',  linewidth=5)
plt.stackplot(days,sleeping,eating,working,playing,color=['m','c','r','k'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('intresting graph\ncheck it out')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
sleeping=[5,6,7,8,4]
eating=  [2,3,4,5,8]
working= [5,6,7,8,9]
playing= [6,7,8,9,4]
smoking=[2,3,4,5,6]
slices=[7,2,2,13,4]
activities=['sleeping','eating','working','playing','smoking']
cols=['c','m','r','b','g']
plt.pie(slices,
        labels=activities,
        colors=cols,
       startangle=90,
       shadow=True,
       explode=(0,0.1,0,0,0),
       autopct='%1.1f%%')
plt.title('intresting graph\ncheck it out')
plt.show()
```
